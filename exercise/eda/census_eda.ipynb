```python
# Check column names for spaces
print("Column names:")
print([f"'{col}'" for col in df.columns])

# Check for leading/trailing spaces in string columns
for col in df.select_dtypes(include=['object']):
    unique_vals = df[col].unique()[:10]  # First 10 unique values
    print(f"\n{col} sample values:")
    for val in unique_vals:
        print(f"'{val}'")  # Quotes will show spaces
```

```python
# The README mentions removing spaces with a text editor
# I try to simulate this to see the difference:
import re

def clean_spaces(text):
    return re.sub(r'\s+', '', text)

# Check a sample of the data
sample_line = open('../data/census.csv').readline()
print(f"Original: {repr(sample_line)}")
print(f"Cleaned:  {repr(clean_spaces(sample_line))}")
```
